# Fuel consumption prediction

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Fuel Consumption Dataset.xlsx')
```

```python
df.head()
```

```python
df.shape
```

```python
x = df[['Engine Size(L)', 'CO2 Emissions(g/km)', 'Cylinders']]  
y = df['Fuel Consumption(Comb (L/100 km))']  
x1 = df['Engine Size(L)']
x2 = df['CO2 Emissions(g/km)']
x3 = df['Cylinders']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state=0)
```

```python
len(x_train)
```

```python
len(y_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
model.fit(x_train, y_train)
```

```python
model.coef_  #beta 1 , beta 2 beta3
```

```python
model.intercept_  #beta0
```

```python
Fuel_Consumption = 4.669247212392522 + 0.05365017*x1 + 0.11073062*x2 + 0.00203697*x3
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
df = pd.DataFrame({'Actual':y_test, 'predicted':y_pred})
df
```
